```python
import torch
import random
from collections import deque
import matplotlib.pyplot as plt
import sys
import os
import numpy as np
%matplotlib inline
```

```python
scores = torch.load('score.pth')
score_2 = []
score_temp = deque(maxlen=100)
for score in scores:
    score_temp.append(score)
    score_2.append(np.average(score_temp))
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(1, len(scores)+1), scores, label = 'Score')
plt.plot(np.arange(1, len(scores)+1), score_2,label = 'Average Score')
plt.legend()
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
```
